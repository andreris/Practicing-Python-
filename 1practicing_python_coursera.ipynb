#### Class 1
- functional programming:
    - lambda: small function to fast transformations.
    - Map: apply a function to each element.
    - Filter: Combine all elements into one result.
    - Reduce: Combine all elements into one result.
    - List: Combine the result into a visible list.
      - maps y filter no devuelven lista solo un iterable

```python
# Lambda  
# 1
lambda x: x * 2
# 2
doble = lambda x: x * 2
print(doble(2))
```

```python
#Map
#forma tradicional
numeros = [1, 2, 3, 4]
resultado = []

for n in numeros:
    resultado.append(n * 2)

print(resultado)
```

```python
#maps
#forma funcional
numeros = [1, 2, 3, 4, 6]
resultado = list(map(lambda x: x + 2, numeros))

print(resultado)
```

```python
#Reduce
from functools import reduce

numeros = [1, 2, 3, 4]

resultado = reduce(lambda x, y: x + y, numeros)

print(resultado)
```

```python
#Filter  
numeros = [1, 2, 3, 4, 5, 6]

resultado = list(filter(lambda x: x % 2 == 0, numeros))

print(resultado)
```

```python
from functools import reduce

vals = ["10", "20", "error", "30"]

def to_int_or_none(s):
    try:
        return int(s)
    except (TypeError, ValueError):
        return None

# Solución funcional por etapas
convertidos = map(to_int_or_none, vals)
validos = filter(lambda x: x is not None, convertidos)
suma = reduce(lambda a, b: a + b, validos, 0)  # 0 como acumulador inicial

print("Suma válidos:", suma)
```

```python
nums = [1, 2, 3, 4, 5, 6]
resultado_tradicional = []
for n in nums:
    triplicado = n * 3
    if triplicado > 10:
        resultado_tradicional.append(triplicado)
print(resultado_tradicional) # Salida: [12, 15, 18]
```

```python
# Usamos map para triplicar y filter para la condición
triplicados = map(lambda x: x * 3, nums)
resultado_funcional = list(filter(lambda x: x > 10, triplicados))
print(resultado_funcional)
```

```python
nombres = [' Ana ', 'luis', ' SOFIA ']
limpios_tradicional = []
for nombre in nombres:
    # strip() quita espacios y upper() pasa a mayúsculas
    nombre_arreglado = nombre.strip().upper()
    limpios_tradicional.append(nombre_arreglado)
print(limpios_tradicional) # Salida: ['ANA', 'LUIS', 'SOFIA']
```

```python
# map aplica la transformación a cada elemento
limpios_funcional = list(map(lambda n: n.strip().upper(), nombres))
print(limpios_funcional)
```

```python
nums = [2, 3, 5]
producto_tradicional = 1 # Empezamos en 1 para multiplicar
for n in nums:
    producto_tradicional = producto_tradicional * n
print(producto_tradicional) # Salida: 30 (2*3*5)
```

```python
from functools import reduce

nums = [2, 3, 5]
# El acumulador 'a' guarda el resultado y 'b' es el siguiente número
producto_funcional = reduce(lambda a, b: a * b, nums)
print(producto_funcional)
```

#### Programacion
- sincronica: espera a que la primera tarea este lista para segur con la siguiente
- asincronica: se ejecuta simultaneamente

```python
import asyncio

async def tarea():
    await asyncio.sleep(2)
    return "Listo"

async def main():
    tareas = [tarea() for _ in range(3)]
    resultados = await asyncio.gather(*tareas)
    print(resultados)

asyncio.run(main())
```
