- String comparison

```python
if word == 'banana':
    print('All right, bananas.')

if word < 'banana':
    print('Your word,' + word + ', comes before banana.')
elif word > 'banana':
    print('Your word,' + word + ', comes after banana.')
else:
    print('All right, bananas.')
```

- String library

- Strings más importantes y usadas en python

  - .lower() : convierte el texto a minúsculas
  - .upper() : convierte el texto a mayúsculas
  - .strip() : elimina espacios al inicio y al final
  - .lstrip() : elimina espacios solo al inicio
  - .rstrip() : elimina espacios solo al final
  - .replace(a, b) : reemplaza una parte del texto por otra
  - .split() : divide el texto en una lista
  - " ".join(lista) : une una lista en un solo texto
  - .find() : devuelve la posición donde aparece algo
  - .count() : cuenta cuántas veces aparece algo
  - .startswith() : verifica si el texto empieza con algo
  - .endswith() : verifica si el texto termina con algo
  - .isdigit() : verifica si el texto contiene solo números
  - .isalpha() : verifica si contiene solo letras
  - .isalnum() : verifica si contiene solo letras y números

```python
len("banana")   # 6 : devuelve la longitud de una cadena
"Hola".lower()   # "hola" elimina las mayúsculas
#-------------------------------------------------------
texto = "HOLA MUNDO"
print(texto.lower())

"hola".upper()   # "HOLA"   agrega mayúsculas
"  hola  ".strip()   # "hola" elimina los espacios a ambos lados
"  hola".lstrip()   # "hola  " elimina los espacios a la izquierda
"hola  ".rstrip()   # "  hola" elimina los espacios a la derecha

"hola mundo".replace("hola", "que tal") #remplaza "mundo" por "Python"
"hola mundo".split() # divide la cadena en una lista de palabras
# ["hola", "mundo"]

" ".join(["hola", "mundo"]) # une los elementos de la lista con un espacio entre ellos
"banana".find("a")   # 1 : busca y devuelve posición
"banana".count("a")   # 3 : devuelve el número de veces que aparece "a"

"hola".startswith("h")   # True : devuelve True si la cadena comienza con "h"
"hola".endswith("a")   # True : devuelve True si la cadena termina con "a"
"123".isdigit()   # True   verifica si contiene solo letras
"hola".isalpha()   # True verifica si contiene letras y  números
```
